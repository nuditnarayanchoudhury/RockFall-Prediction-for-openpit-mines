```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../dataset/synthetic_drone_dataset.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
df.hist(figsize=(10, 8), bins=20)
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Feature Boxplots")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['crack_density'], bins=30, kde=True, color='red')
plt.title("Crack Density Distribution")
plt.xlabel("Crack Density")
plt.ylabel("Frequency")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))

sns.histplot(df['debris_entropy'], bins=30, kde=True, color='blue', ax=axes[0])
axes[0].set_title("Debris Entropy")

sns.histplot(df['debris_contrast'], bins=30, kde=True, color='green', ax=axes[1])
axes[1].set_title("Debris Contrast")

sns.histplot(df['debris_homogeneity'], bins=30, kde=True, color='purple', ax=axes[2])
axes[2].set_title("Debris Homogeneity")

plt.suptitle("Debris Feature Distributions", fontsize=14)
plt.tight_layout()
plt.show()
```
