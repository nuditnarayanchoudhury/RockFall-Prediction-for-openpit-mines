```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "../dataset/refined_sih_dataset_ML_ready.csv" 
df = pd.read_csv(file_path)
```

```python
print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
key_features = ["rainfall", "Displacement_mm", "Strain_micro", "PorePressure_kPa", "SeismicVibration_mm/s"]

for col in key_features:
    if col in df.columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(df[col], bins=50, kde=True)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()
```

```python
for col in key_features:
    if col in df.columns:
        plt.figure(figsize=(8, 5))
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()
```

```python
if all(col in df.columns for col in key_features):
    sns.pairplot(df[key_features].sample(1000, random_state=42)) 
    plt.suptitle("Pairwise Feature Relationships (Sampled)", y=1.02)
    plt.show()
```

```python
if "YEAR" in df.columns and "rainfall" in df.columns:
    plt.figure(figsize=(12, 6))
    plt.plot(df["YEAR"], df["rainfall"], label="Rainfall", color="blue")
    plt.xlabel("Year")
    plt.ylabel("Rainfall")
    plt.title("Rainfall Trend Over Years")
    plt.legend()
    plt.show()
```
